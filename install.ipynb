# install on a macbook (notes, do not run)

install pyenv for rootless python lightweight env management

```python
curl https://pyenv.run | bash
```

update zsh

```python
echo 'export PYENV_ROOT="$HOME/.pyenv"' >> ~/.zshrc
echo 'command -v pyenv >/dev/null || export PATH="$PYENV_ROOT/bin:$PATH"' >> ~/.zshrc
echo 'eval "$(pyenv init -)"' >> ~/.zshrc
source ~/.zshrc
```

install python 3.12 (3.14 likely too recent)

```python
pyenv install 3.12.8
pyenv local 3.12.8
python --version

# Tkinter will not be installe but it’s OK
```

install scvi in a venv

```python
python -m venv scvi-env
source scvi-env/bin/activate
pip install --upgrade pip setuptools wheel

pip install jupyterlab scvi-tools ipykernel leidenalg

# and the kernel …

python -m ipykernel install --user --name scvi-env --display-name "Python (scvi)"
```

start the lab

```python
jupyter lab
```
